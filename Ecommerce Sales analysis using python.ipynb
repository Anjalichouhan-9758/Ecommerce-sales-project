```python
# Merge files
```

```python
amazon.rename(columns = {'SKU':'sku'},inplace = True)
sale.columns
# amazon.columns
```

```python
sale.rename(columns = {'sku_code':'sku'},inplace = True)
```

```python
sales_analysis = pd.merge(amazon, sale, left_on='sku',right_on = 'sku', how='left')
```

```python
# Step - 2 Merge Amazon + Sale
sales_analysis = pd.merge(amazon, sale, on ='sku', how ='left')
```

```python
sales_analysis.shape
```

```python
sales_analysis.head()
```

```python
sales_analysis.isnull().sum()
```

```python
(sales_analysis['category_x'] == sales_analysis['category_y']).value_counts()
```

```python
(sales_analysis['size_x'] == sales_analysis['size_y']).value_counts()
```

```python
sales_analysis.drop(['category_y'], axis=1, inplace=True)
sales_analysis.rename(columns={'category_x':'category'}, inplace=True)
```

```python
sales_analysis['size'] = sales_analysis['size_x'].combine_first(sales_analysis['size_y'])
sales_analysis.drop(['size_x','size_y'], axis=1, inplace=True)
```

```python
sales_analysis['design_no.'] = sales_analysis['design_no.'].fillna("Unknown")
sales_analysis['category'] = sales_analysis['category'].fillna("Unknown")
sales_analysis['size'] = sales_analysis['size'].fillna("Unknown")
sales_analysis['color'] = sales_analysis['color'].fillna("Unknown")

sales_analysis['stock'] = sales_analysis['stock'].fillna(0)
```
